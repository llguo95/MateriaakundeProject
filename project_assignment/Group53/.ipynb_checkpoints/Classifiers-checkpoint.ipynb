```python
# 3D dataset with 4 different classifiers coilable, yields and non coilable
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold

kf = KFold(n_splits=4)
data = pd.read_pickle('DoE_results_3d_1k.pkl')
data_df = data['points']
feature_names = data_df.columns.values[0:3] # not including the "ID" column

X = data_df.iloc[:, 0:3 ].values  #get all the 3d variables

y_string = data_df.loc[:, 'coilable'].values
y = pd.Categorical(pd.factorize(y_string)[0])
target_names = data_df.loc[:, 'coilable'].unique()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0)

# Parameters
n_classes = 3
plot_colors = "ryb"
plot_step = 0.02
C = 1.0 # SVM regularization parameter
svm_model = svm.SVC(kernel='rbf', gamma=0.7, C=C)
XG_model=xgb.XGBClassifier(random_state=1,learning_rate=0.01)

fig, axs = plt.subplots(4,3,figsize=(15, 15)) # open figure with larger size
plt.suptitle("Decision surface of a Decision Tree, Random Forrest and SVM using paired features")

print('Accuracy for different Classifiers')
print('DT \tRF \tSVM \tXG \tFeature')

for pairidx, pair in enumerate([[0, 1], [0, 2], [1, 2]]):

    X_temp = X_train[:, pair]
    X_temp_test = X_test[:, pair] # The correct test data to validate prediction
    
    # Train
    clfDT = DecisionTreeClassifier().fit(X_temp, y_train) # Decision Tree
    clfRF = RandomForestClassifier(max_depth=2, random_state=0).fit(X_temp, y_train) # Random Forrest
    svm_model.fit(X_temp, y_train)  # SVM
    XG_model.fit(X_temp, y_train)   # XG Boost

    # Set the scale correctly
    x1_min, x1_max = X_temp[:, 0].min() - 0, X_temp[:, 0].max() + 0
    x2_min, x2_max = X_temp[:, 1].min() - 0, X_temp[:, 1].max() + 0
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, plot_step),
                         np.arange(x2_min, x2_max, plot_step))

    # Predict using the different Classifiers
    YDT = clfDT.predict(np.c_[xx1.ravel(), xx2.ravel()])  # Decision Tree
    YRF = clfRF.predict(np.c_[xx1.ravel(), xx2.ravel()])  # Random Forrest
    YSVM = svm_model.predict(np.c_[xx1.ravel(), xx2.ravel()]) #SVM
    YXG = XG_model.predict(np.c_[xx1.ravel(), xx2.ravel()])   # XG Boost
    print(clfDT.score(X_temp_test, y_test),'\t',clfRF.score(X_temp_test, y_test),'\t',svm_model.score(X_temp_test, y_test),'\t',XG_model.score(X_temp_test,y_test),'\t',feature_names[pair[0]],'/', feature_names[pair[1]])
    YDT = YDT.reshape(xx1.shape)
    YRF = YRF.reshape(xx1.shape)
    YSVM = YSVM.reshape(xx1.shape)
    YXG = YXG.reshape(xx1.shape)

    # Plot the decision boundary
    axs[0, pairidx].contourf(xx1, xx2, YDT, cmap=cm.RdYlBu)
    axs[0, pairidx].set(xlabel=feature_names[pair[0]],ylabel=feature_names[pair[1]])
    axs[1, pairidx].contourf(xx1, xx2, YRF, cmap=cm.RdYlBu)
    axs[1, pairidx].set(xlabel=feature_names[pair[0]],ylabel=feature_names[pair[1]])
    axs[2, pairidx].contourf(xx1, xx2, YSVM, cmap=cm.RdYlBu)
    axs[2, pairidx].set(xlabel=feature_names[pair[0]],ylabel=feature_names[pair[1]])
    axs[3, pairidx].contourf(xx1, xx2, YXG, cmap=cm.RdYlBu)
    axs[3, pairidx].set(xlabel=feature_names[pair[0]],ylabel=feature_names[pair[1]])

    # Plot the training points
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y_train == i)
        for j in range(4):
          axs[j, pairidx].scatter(X_temp[idx, 0], X_temp[idx, 1], c=color,
                    label=target_names[i],cmap=plt.cm.RdYlBu,
                    edgecolor='black', s=15)

fig.tight_layout(h_pad=2, w_pad=5, pad=5)
plt.legend(loc='lower right', borderpad=0, handletextpad=0)
```

```python
# 7D dataset with 2 different classifiers coilable, yields and non coilable
import pickle
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_fscore_support

data = pd.read_pickle('DoE_results_7d_50k.pkl')
data_df = data['points']

feature_names = data_df.columns.values[0:7] # not including the "ID" column
X = data_df.iloc[:, 0:7 ].values  #get all the 7d variables
y_string = data_df.loc[:, 'coilable'].values
y = pd.Categorical(pd.factorize(y_string)[0])
target_names = data_df.loc[:, 'coilable'].unique()

data_dmatrix = xgb.DMatrix(data=X,label=y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)

dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=feature_names)

# create XG Boost
paramsXG = {"objective":"reg:logistic",'colsample_bytree': 0.3,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}

XG_class=xgb.XGBClassifier(objective ='reg:logistic', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)

C = 1.0 # SVM regularization parameter
svm_model = svm.SVC(kernel='rbf', gamma=0.7, C=C)

# Train all classifiers
XG_class.fit(X_train,y_train)       #XG Boost
clfDT = DecisionTreeClassifier().fit(X_train, y_train)   # Decision Tree
clfRF = RandomForestClassifier(max_depth=2, random_state=0).fit(X_train, y_train) # Random Forrest
svm_model.fit(X_train, y_train)     # SVM

# Predict using the different Classifiers
YDT = clfDT.predict(X_test)         # Decision Tree
YRF = clfRF.predict(X_test)         # Random Forrest
YSVM = svm_model.predict(X_test)    # SVM
predsXG = XG_class.predict(X_test)  # XG Boost

print('\t DT\t RF\t SVM\t XG')
print('Score','\t',clfDT.score(X_test, y_test),'\t',clfRF.score(X_test, y_test),'\t',svm_model.score(X_test, y_test),'\t', XG_class.score(X_test,y_test))
print('MSE:','\t',mean_squared_error(y_test,YDT),'\t', mean_squared_error(y_test,YRF),'\t', mean_squared_error(y_test,YSVM),'\t', mean_squared_error(y_test,predsXG))
print('F1:','\t',f1_score(y_test,YDT),'\t',f1_score(y_test,YRF),'\t',f1_score(y_test,YSVM),'\t',f1_score(y_test,predsXG))

print('\t pre\t rec\t f1\t sup')
print('DT:\t', precision_recall_fscore_support(y_test, YDT, average='weighted'))
print('RF:\t', precision_recall_fscore_support(y_test, YRF, average='weighted'))
print('SVM:\t', precision_recall_fscore_support(y_test, YSVM, average='weighted'))
print('XG:\t', precision_recall_fscore_support(y_test, predsXG, average='weighted'))

cv_results = xgb.cv(dtrain=data_dmatrix, params=paramsXG, nfold=5,
                    num_boost_round=1000,early_stopping_rounds=100,metrics="rmse", as_pandas=True, seed=123)
cv_results.head()
print((cv_results["test-rmse-mean"]).tail(1))
print(XG_class.score(X_test,y_test))


xgb.plot_tree(XG_class, num_trees=0)
plt.rcParams['figure.figsize'] = [50, 10]
plt.show()

xgb.plot_importance(XG_class)
plt.rcParams['figure.figsize'] = [5, 5]
plt.show()
```

```python
# 3D dataset with 2 different classifiers coilable and non coilable
import pickle
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_fscore_support

data = pd.read_pickle('DoE_results_3d_1k.pkl')
data_df = data['points']

feature_names = data_df.columns.values[0:3] # not including the "ID" column
X = data_df.iloc[:, 0:3 ].values  #get all the 3d variables
y_string = data_df.loc[:, 'coilable'].values
y_string[y_string == 'coilable (but yields)'] = 'coilable'
y = pd.Categorical(pd.factorize(y_string)[0])
target_names = data_df.loc[:, 'coilable'].unique()

data_dmatrix = xgb.DMatrix(data=X,label=y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)

dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=feature_names)

# XG Boost
paramsXG = {"objective":"reg:logistic",'colsample_bytree': 0.3,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}

XG_class=xgb.XGBClassifier(objective ='reg:logistic', colsample_bytree = 0.3, learning_rate = 0.05,
                max_depth = 10, alpha = 10, n_estimators = 10)

C = 1.0 # SVM regularization parameter
svm_model = svm.SVC(kernel='rbf', gamma=0.7, C=C)

# Train
XG_class.fit(X_train,y_train)     #XG Boost
clfDT = DecisionTreeClassifier().fit(X_train, y_train)   # Decision Tree
clfRF = RandomForestClassifier(max_depth=2, random_state=0).fit(X_train, y_train) # Random Forrest
svm_model.fit(X_train, y_train)    # SVM

# Predict using the different Classifiers
YDT = clfDT.predict(X_test)         # Decision Tree
YRF = clfRF.predict(X_test)         # Random Forrest
YSVM = svm_model.predict(X_test)    # SVM
predsXG = XG_class.predict(X_test)  # XG Boost

print('\t DT\t RF\t SVM\t XG')
print('Score','\t',clfDT.score(X_test, y_test),'\t',clfRF.score(X_test, y_test),'\t',svm_model.score(X_test, y_test),'\t', XG_class.score(X_test,y_test))
print('MSE:','\t',mean_squared_error(y_test,YDT),'\t', mean_squared_error(y_test,YRF),'\t', mean_squared_error(y_test,YSVM),'\t', mean_squared_error(y_test,predsXG))
print('F1:','\t',f1_score(y_test,YDT),'\t',f1_score(y_test,YRF),'\t',f1_score(y_test,YSVM),'\t',f1_score(y_test,predsXG))
print('\t pre\t rec\t f1\t sup')
print('DT:\t', precision_recall_fscore_support(y_test, YDT, average='weighted'))
print('RF:\t', precision_recall_fscore_support(y_test, YRF, average='weighted'))
print('SVM:\t', precision_recall_fscore_support(y_test, YSVM, average='weighted'))
print('XG:\t', precision_recall_fscore_support(y_test, predsXG, average='weighted'))

cv_results = xgb.cv(dtrain=data_dmatrix, params=paramsXG, nfold=5,
                    num_boost_round=1000,early_stopping_rounds=100,metrics="rmse", as_pandas=True, seed=123)
cv_results.head()
print(((cv_results["test-rmse-mean"]).tail(1))**2)

xgb.plot_tree(XG_class, num_trees=0)
plt.rcParams['figure.figsize'] = [50, 10]
plt.show()

xgb.plot_importance(XG_class)
plt.rcParams['figure.figsize'] = [5, 5]
plt.show()
```

```python
# 3D dataset with 3 different classifiers coilable, yields and non coilable
# different number of training points
import pickle
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_recall_fscore_support
from matplotlib.legend_handler import HandlerLine2D

data = pd.read_pickle('DoE_results_3d_1k.pkl')
data_df = data['points']

feature_names = data_df.columns.values[0:3] # not including the "ID" column
X = data_df.iloc[:, 0:3 ].values  #get all the 3d variables
y_string = data_df.loc[:, 'coilable'].values
y = pd.Categorical(pd.factorize(y_string)[0])

test_splits = np.linspace(0.1, 0.9, 9, endpoint=True)

scores_train = []
scores_test = []
print('\t pre\t rec\t f1\t sup')

for split in test_splits:
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-split, random_state=123)

  # Train
  clfDT = DecisionTreeClassifier().fit(X_train, y_train)   # Decision Tree

  # Predict using Decision Tree
  YDT = clfDT.predict(X_test)         # Decision Tree
  
  #Calculate scores
  scoreTest = precision_recall_fscore_support(y_test, YDT, average='weighted')
  avgTest = (scoreTest[0] + scoreTest[1] + scoreTest[2]) / 3

  # Add score to respecive scores
  scores_test.append(avgTest)

  print(1-split,'%\t', scoreTest)

line1, = plt.plot(test_splits*100, scores_test, 'b', label="Avg Test Score")

plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})

plt.ylabel('Score')
plt.xlabel('Percentage of data for Training points')
plt.show()
```

```python
# 3D dataset with 3 different classifiers coilable, yields and non coilable
# Hyperparameters influence
import pickle
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_recall_fscore_support
from matplotlib.legend_handler import HandlerLine2D

data = pd.read_pickle('DoE_results_3d_1k.pkl')
data_df = data['points']

feature_names = data_df.columns.values[0:3] # not including the "ID" column
X = data_df.iloc[:, 0:3 ].values  #get all the 3d variables
y_string = data_df.loc[:, 'coilable'].values
y = pd.Categorical(pd.factorize(y_string)[0])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)

# Max Tree Depth
max_depths = np.linspace(1, 32, 32, endpoint=True)
scores0 = []
scores1 = []

for max_depth in max_depths:
  dt = DecisionTreeClassifier(max_depth=max_depth)
  dt.fit(X_train, y_train)
  train_pred = dt.predict(X_train)
  trest_pred = dt.predict(X_test)

  #Calculate scores
  scoreTrain = precision_recall_fscore_support(y_train, train_pred, average='weighted')
  scoreTest = precision_recall_fscore_support(y_test, trest_pred, average='weighted')
  avgTrain = (scoreTrain[0] + scoreTrain[1] + scoreTrain[2]) / 3
  avgTest = (scoreTest[0] + scoreTest[1] + scoreTest[2]) / 3

  # Add score to respecive scores
  scores0.append(avgTrain)
  scores1.append(avgTest)

line1, = plt.plot(max_depths, scores0, 'b', label="Avg Train Score")
line2, = plt.plot(max_depths, scores1, 'r', label="Avg Test Score")

plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})

plt.ylabel('Score')
plt.xlabel('Tree depth')
plt.show()

# Min Sample Splits
min_samples_splits = np.linspace(0.1, 1.0, 25, endpoint=True)
scores0 = []
scores1 = []

for min_samples_split in min_samples_splits:
  dt = DecisionTreeClassifier(min_samples_split=min_samples_split)
  dt.fit(X_train, y_train)
  train_pred = dt.predict(X_train)
  trest_pred = dt.predict(X_test)

  #Calculate scores
  scoreTrain = precision_recall_fscore_support(y_train, train_pred, average='weighted')
  scoreTest = precision_recall_fscore_support(y_test, trest_pred, average='weighted')
  avgTrain = (scoreTrain[0] + scoreTrain[1] + scoreTrain[2]) / 3
  avgTest = (scoreTest[0] + scoreTest[1] + scoreTest[2]) / 3

  # Add score to respecive scores
  scores0.append(avgTrain)
  scores1.append(avgTest)

line1, = plt.plot(min_samples_splits, scores0, 'b', label="Avg Train Score")
line2, = plt.plot(min_samples_splits, scores1, 'r', label="Avg Test Score")

plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})

plt.ylabel('Score')
plt.xlabel('Min Sample Splits')
plt.show()


# Min Smaple Leafs
min_samples_leafs = np.linspace(0.01, 0.5, 25, endpoint=True)
scores0 = []
scores1 = []

for min_samples_leaf in min_samples_leafs:
  dt = DecisionTreeClassifier(min_samples_leaf=min_samples_leaf)
  dt.fit(X_train, y_train)
  train_pred = dt.predict(X_train)
  trest_pred = dt.predict(X_test)

  #Calculate scores
  scoreTrain = precision_recall_fscore_support(y_train, train_pred, average='weighted')
  scoreTest = precision_recall_fscore_support(y_test, trest_pred, average='weighted')
  avgTrain = (scoreTrain[0] + scoreTrain[1] + scoreTrain[2]) / 3
  avgTest = (scoreTest[0] + scoreTest[1] + scoreTest[2]) / 3

  # Add score to respecive scores
  scores0.append(avgTrain)
  scores1.append(avgTest)

line1, = plt.plot(min_samples_leafs, scores0, 'b', label="Avg Train Score")
line2, = plt.plot(min_samples_leafs, scores1, 'r', label="Avg Test Score")

plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})

plt.ylabel('Score')
plt.xlabel('Min Sample Splits')
plt.show()
```

```python
# Train baseline
base = DecisionTreeClassifier().fit(X_train, y_train)   # Decision Tree

# Predict using Decision Tree
Ybase = base.predict(X_test)         # Decision Tree

# Hyperparamter optimization
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_dist = {"max_depth": [3, None],
              "min_samples_split": randint(1, 9),
              "min_samples_leaf": randint(1, 9),
              "criterion": ["gini", "entropy"]}

hyper_tree = DecisionTreeClassifier()
hyper_tree_cv = RandomizedSearchCV(hyper_tree, param_dist, cv=100)

hyper_tree_cv.fit(X_train, y_train)

Yhyper = hyper_tree_cv.predict(X_test)

print('\t\t precision \t\t recall \t f1')
print('Baseline\t', precision_recall_fscore_support(y_test, Ybase, average='weighted'))
print('Optimized\t', precision_recall_fscore_support(y_test, Yhyper, average='weighted'))
print("Tuned Decision Tree Parameters: {}".format(hyper_tree_cv.best_params_))
print("Best score is {}".format(hyper_tree_cv.best_score_))
```
