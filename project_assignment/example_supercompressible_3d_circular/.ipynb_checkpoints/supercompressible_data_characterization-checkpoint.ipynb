# Supercompressible: data characterization

This notebook intends to characterized the data presented in the dataset. Visualization tools will be used whenever possible.

So, let's start by importing the required libraries:

```python
# standard library
import pickle
import itertools

# third-party
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

Now, let's open the file and store all the information in the Python environment.

```python
filename = 'DoE_results.pkl'
with open(filename, 'rb') as file:
    data = pickle.load(file)
    
print(data)
```

```python
# define number of points to start with
N = 1000
points = data['points'].loc[:(N-1)]
print(points)
```

To make sure all the data is properly labeled in the plots, let's create meaningful labels. To make it easy to use, each variable is represented by a tuple, where the first component is the name of the variable in the dataset and the second component corresponds to its label and the third to its bounds.

```python
# inputs
input_vars = [('ratio_d', '$\\frac{d}{D_1}$'), ('ratio_pitch', '$\\frac{P}{D}$'), ('ratio_top_diameter', '$\\frac{D_1 - D_2}{D_1}$')]
for i, input_var in enumerate(input_vars):
    input_vars[i] = input_var + (data['doe_variables'][input_var[0]],)

print(input_vars)
    
# outputs
output_vars = [('coilable', 'coilable'), ('sigma_crit', '$\\sigma_{crit}$'), ('energy', '$E_{abs}$')]
```

Let's forget the outputs for a while and take a look to the input variables.

```python
# get number of inputs
n_inputs = len(points.columns) - 3

# get X data
X = points.iloc[:, range(n_inputs)].values

print('X:', X)
print('X shape:', np.shape(X))
```

Let's start bt plotting the variables in pairs using scatter plots.

```python
# get all possible combinations
combinations = list(itertools.combinations(range(n_inputs), 2))

for comb in combinations:
    i, j = comb
    xx = X[:, i]
    yy = X[:, j]
    
    plt.figure()
    plt.scatter(xx, yy, s=1)
    plt.xlabel(input_vars[i][1])
    plt.ylabel(input_vars[j][1])
    plt.xlim(input_vars[i][2][0], input_vars[i][2][1])
    plt.ylim(input_vars[j][2][0], input_vars[j][2][1])
```

These plots are not useful, but reflect the way the dataset was generated (the space is densely and uniformly filled). Let's take a look to the first $n$ points.

```python
n = 500

for comb in combinations:
    i, j = comb
    xx = X[:n, i]
    yy = X[:n, j]
    
    plt.figure()
    plt.scatter(xx, yy, s=10)
    plt.xlabel(input_vars[i][1])
    plt.ylabel(input_vars[j][1])
    plt.xlim(input_vars[i][2][0], input_vars[i][2][1])
    plt.ylim(input_vars[j][2][0], input_vars[j][2][1])
```

The following is a more condensed representation of the same information.

```python
n = 100
i, j, k = 1, 2, 0
xx = X[:n, i]
yy = X[:n, j]
zz = X[:n, k]

plt.figure()
plt.scatter(xx, yy, s=20, c=zz, cmap=plt.cm.coolwarm)
plt.xlabel(input_vars[i][1])
plt.ylabel(input_vars[j][1])
plt.xlim(input_vars[i][2][0], input_vars[i][2][1])
plt.ylim(input_vars[j][2][0], input_vars[j][2][1])
clb = plt.colorbar()
clb.ax.set_title(input_vars[k][1])
plt.show()
```

## Coilability

Let's now use the classification variable and verify if we can draw any conclusion.

```python
y = points.loc[:, 'coilable'].values
n_categs = len(np.unique(y))
categs = [(0, 'not coilable'), (1, 'coilable'), (2, 'coilable (but yields)')]
if n_categs < 3:
    categs = categs[:2]

colors = ['r', 'g', 'b'][:n_categs]
```

```python
n = 10000

for comb in combinations:
    i, j = comb
    fig, ax = plt.subplots()
    for color, categ in zip(colors, categs):
        indices = np.where(y[:n] == categ[0])
        xx = X[:n, i][indices]
        yy = X[:n, j][indices]
        ax.scatter(xx, yy, s=1, color=color, label=categ[1])
        
    ax.set_xlabel(input_vars[i][1])
    ax.set_ylabel(input_vars[j][1])
    ax.set_xlim(input_vars[i][2][0], input_vars[i][2][1])
    ax.set_ylim(input_vars[j][2][0], input_vars[j][2][1])
    ax.legend(bbox_to_anchor=(1, 1))
```

From the pictures above, notice that the categorization problem can be done with high sucess with two features (properly choosen). The distinction between *coilable* and *not coilable* is easier done than the distinction between *coilable* and *coilable (but yields)*. A sequential classification may be the best strategy to follow, i.e. start to separate *coilable* from *not coilable* and then, if *coilable*, classify between *coilable* and *coilable (but yields)*.

## Energy absorbed

Let's now look to the energy absorbed.

```python
var_name = 'energy'
```

```python
# missing indices
indices = pd.notnull(points.loc[:, var_name]).values

# get y data
y = points.loc[indices, var_name].values.copy()

# get X data
X = points.iloc[indices, range(n_inputs)].values.copy()

print('X shape:', np.shape(X))
print('y shape:', np.shape(y))
```

```python
fig1, ax1 = plt.subplots()
ax1.hist(y);
ax1.set_xlabel('Energy')
ax1.set_ylabel('Frequency');
```

This histogram shows the existence of outliers. Let's get rid of them.

```python
def perform_cleaning(X, y, n_std_cleaning):
    
    # compute threshold
    y_mean, y_std = np.mean(y), np.std(y)
    y_thresh = y_mean + n_std_cleaning * y_std

    # indices
    print(y_thresh)
    indices = np.where(y < y_thresh)
    n_dismissed_points = len(y) - len(indices[0])
    print('Dismissed points:', n_dismissed_points)

    # X and y
    y = y[indices]
    X = X[indices]
    print(np.shape(y))
    print(np.shape(X))

    return X, y, n_dismissed_points
```

```python
n_std_cleaning = 5  # dismiss points that fall outside this range

n_dismissed_points = 1
n_max_iter = 2
n_iter = 0
while n_dismissed_points > 0 and n_iter < n_max_iter:
    X, y, n_dismissed_points = perform_cleaning(X, y, n_std_cleaning)
    n_iter += 1
```

```python
fig1, ax1 = plt.subplots()
ax1.hist(y);
ax1.set_xlabel('Energy')
ax1.set_ylabel('Frequency');
```

The histogram is now more meaningful. The application of the Machine Learning algorithms helps to define the strategy of cleaning: it clearly shows that points with energy above 90 are outliers (erros in the computation of the energy).

Let's now find the best solution in the database for the energy.

```python
y_coil = points.loc[:, 'coilable'].values.copy()
y_energy = points.loc[:, 'energy'].values.copy()
X_complete = points.iloc[:, range(n_inputs)].values.copy()

indices = np.where((y_coil == 1) & (~np.isnan(y_energy)))
yy_valid = y_energy[indices]
xx_valid = X_complete[indices]
max_index, max_energy = np.argmax(yy_valid), np.max(yy_valid)

print('Max energy:', max_energy)
print('Corresponding data point:', xx_valid[max_index])
```
